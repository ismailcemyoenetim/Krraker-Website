```python
import pandas as pd

# Read the CSV file from desktop
df = pd.read_csv('~/Desktop/lichess_db_puzzle.csv')

# The CSV columns will automatically map to these variables:
# PuzzleId, FEN, Moves, Rating, RatingDeviation, Popularity, NbPlays, Themes, GameUrl, OpeningTags
```

```python
df.head()
```

```python
# Filter puzzles between 2000-2300 rating
filtered_df = df[(df['Rating'] >= 2000) & (df['Rating'] <= 2300)]

# Display first few rows of filtered data
print("Number of puzzles between 2000-2300 rating:", len(filtered_df))
print("\nFirst few filtered puzzles:")
print(filtered_df.head())

# Export to JSON file
filtered_df.to_json('puzzles_2000_2300.json', orient='records')
print("\nFiltered puzzles have been exported to 'puzzles_2000_2300.json'")
```

```python
# Filter puzzles between 2000-2300 rating and take only first 25
filtered_df = df[(df['Rating'] >= 2000) & (df['Rating'] <= 2300)].head(25)

# Display number of puzzles
print("Number of puzzles exported:", len(filtered_df))

# Export to JSON file
filtered_df.to_json('puzzles_2000_2300.json', orient='records')
print("\nFiltered puzzles have been exported to 'puzzles_2000_2300.json'")
```
